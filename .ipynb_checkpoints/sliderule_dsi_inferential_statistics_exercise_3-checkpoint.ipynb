# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

### Setup an appropriate hypothesis test.

Null Hypothesis : No significant difference between average excess readmission ratios of large and small hospitals.

Alternate Hypothesis : There is significant difference between average excess readmission ratios of large and small hospitals.


### Compute and report the observed significance value (or p-value).

From the preliminary report above:

> B. Statistics
    - In hospitals/facilities with number of discharges < 100, mean excess readmission rate is 1.023 and 63% have excess readmission rate greater than 1 
    - In hospitals/facilities with number of discharges > 1000, mean excess readmission rate is 0.978 and 44% have excess readmission rate greater than 1 
    
Using their parameters for large and small hospitals, we will split our data and perform a two-sample t-test.

```python
sm_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'].astype(int) <= 100]
lg_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'].astype(int) >= 1000]

group_sm = sm_hosp['Excess Readmission Ratio'][81:-3]
group_lg = lg_hosp['Excess Readmission Ratio'][81:-3]

print('Small Hospital - Standard Deviation: {}'.format(np.std(group_sm)))
print('Large Hospital - Standard Deviation: {}'.format(np.std(group_lg)))
```

```python
# perform two-sample t-test
scipy.stats.ttest_ind(group_sm,group_lg, equal_var=False)
```

```python
mean_sm_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'].astype(int) <= 364]
mean_lg_hosp = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'].astype(int) > 364]

mean_group_sm = mean_sm_hosp['Excess Readmission Ratio'][81:-3]
mean_group_lg = mean_lg_hosp['Excess Readmission Ratio'][81:-3]

print('Small Hospital - Standard Deviation: {}'.format(np.std(mean_group_sm)))
print('Large Hospital - Standard Deviation: {}'.format(np.std(mean_group_lg)))
```

```python
# perform two-sample t-test
scipy.stats.ttest_ind(avg_group_sm,avg_group_lg, equal_var=False)
```

### Report statistical significance for α = .01. 

By simply recreating their analysis, we can see that it is significant at the α = .01 level. When we split the the hospitals by 364, the mean number of discharges, we do see a higher t-score and more significant p-value. ### Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?

#### Cohen's D effect size:
> Cohen classified effect sizes as small (d = 0.2), medium (d = 0.5), and large (d ≥ 0.8). According to Cohen, “a medium effect of .5 is visible to the naked eye of a careful observer.

```python
diff = sm_hosp['Excess Readmission Ratio'][81:-3].mean() - lg_hosp['Excess Readmission Ratio'][81:-3].mean()

n1, n2 = len(sm_hosp['Excess Readmission Ratio'][81:-3]), len(lg_hosp['Excess Readmission Ratio'][81:-3])
var1 = sm_hosp['Excess Readmission Ratio'][81:-3].var()
var2 = lg_hosp['Excess Readmission Ratio'][81:-3].var()

s_sq = (n1 * var1 + n2 * var2) / (n1 + n2)
d = diff / np.sqrt(s_sq)
print('Cohen\'s D: {:2f}'.format(d))
```

```python
mean_diff = mean_sm_hosp['Excess Readmission Ratio'][81:-3].mean() - mean_lg_hosp['Excess Readmission Ratio'][81:-3].mean()

mean_n1, mean_n2 = len(mean_sm_hosp['Excess Readmission Ratio'][81:-3]), len(mean_lg_hosp['Excess Readmission Ratio'][81:-3])
mean_var1 = mean_sm_hosp['Excess Readmission Ratio'][81:-3].var()
mean_var2 = mean_lg_hosp['Excess Readmission Ratio'][81:-3].var()

mean_s_sq = (mean_n1 * mean_var1 + mean_n2 * mean_var2) / (mean_n1 + mean_n2)
mean_d = mean_diff / np.sqrt(mean_s_sq)
print('Cohen\'s D: {:2f}'.format(mean_d))
```
